```python
plt.subplot(1,2,1)
plt.plot(x,z,'y+-')
plt.title("X-Z Plot")
plt.subplot(1,2,2)
plt.plot(z,x,'b*-')
plt.title("Z-X Plot")
```

```python
x1 = np.linspace(0.0, 5.0) 
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) 
y2 = np.cos(2 * np.pi * x2)
plt.subplot(2,1,1)
plt.scatter(x1,y1)
plt.plot(x1,y1)
plt.ylabel("Damped Oscillation")
plt.title("A table of 2 supblots")
plt.subplot(2,1,2)
plt.scatter(x2,y2)
plt.plot(x2,y2)
plt.ylabel("Undamped")
plt.xlabel("time(s)")
```
