```python
import copy #for copying the array
import time
from prettytable import PrettyTable
```

```python
def table(field,grid): #for making a beautiful table of the grid
    x = PrettyTable()
    x.field_names = [field,""]
    line1 = str(grid[0]) + " " +  str(grid[1]) + " " +  str(grid[2])
    line2 = str(grid[3]) + " " +  str(grid[4]) + " " +  str(grid[5])
    line3 = str(grid[6]) + " " +  str(grid[7]) + " " +  str(grid[8])
    line4 = "Moves: " + str(grid[9:])
    x.add_row([line1,""])
    x.add_row([line2,""])
    x.add_row([line3,""])
    x.add_row([line4,""])
    return x
```

```python
def printGrid(grid): #for printing the table
    global case
    field = "Case: " + str(case)
    print(table(field,grid))
    print()
    case += 1
```

```python
def checkFinal(grid):
    if grid[:9]==final:
        printGrid(grid)
        return True
    if grid[:9] not in visited:
        printGrid(grid)
        queue.append(grid)
        visited.append(grid[:9])
    return False
```

```python
print("Please enter value in this form:")
print("a b c")
print("d e f")
print("g h i")
print()
print("Enter start node. Enter zero instead of the blank space:")
start1 = input().split()
start2 = input().split()
start3 = input().split()
print()
print("Enter goal node. Enter zero instead of the blank space:")
end1 = input().split()
end2 = input().split()
end3 = input().split()
start = [int(start1[0]),int(start1[1]),int(start1[2]), int(start2[0]),int(start2[1]),int(start2[2]), int(start3[0]),int(start3[1]),int(start3[2])]
final = [int(end1[0]),int(end1[1]),int(end1[2]), int(end2[0]),int(end2[1]),int(end2[2]), int(end3[0]),int(end3[1]),int(end3[2])]
    
found = False
case = 0
visited = []
queue = []
queue.append(start)
visited.append(start)
printGrid(start)
t0 = time.time()
while (not found):
    current = queue.pop(0)
    blank = current.index(0)
    if blank!=0 and blank!=1 and blank!=2:
        up = copy.deepcopy(current)
        up[blank] = up[blank-3]
        up[blank-3] = 0
        up.append('up')
        found = checkFinal(up)
    if blank!=0 and blank!=3 and blank!=6 and found==False:
        left = copy.deepcopy(current)
        left[blank] = left[blank-1]
        left[blank-1] = 0
        left.append('left')
        found = checkFinal(left)
    if blank!=6 and blank!=7 and blank!=8 and found==False:
        down = copy.deepcopy(current)
        down[blank] = down[blank+3]
        down[blank+3] = 0
        down.append('down')
        found = checkFinal(down)
    if blank!=2 and blank!=5 and blank!=8 and found==False:
        right = copy.deepcopy(current)
        right[blank] = right[blank+1]
        right[blank+1] = 0
        right.append('right')
        found = checkFinal(right)

t1 = time.time()
print("Time: " + str(t1-t0) + " seconds")
```
